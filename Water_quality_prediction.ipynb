```python
#Importing needed modules.
from pandas import read_csv , DataFrame , concat
from sklearn.preprocessing import LabelEncoder , StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
#Instancing from 'LabelEncoder' and 'StandardScaler' objects.
encoder = LabelEncoder()
scaler = StandardScaler()
```

```python
#Reading csv file and getting 10 sample of that randomly.
Dir = "D:\\datasets\\water_quality_heavy\\Water Quality Prediction.csv"
file = read_csv(Dir)
df = file.copy()
df.sample(10)
```

```python
#Dropping unuseful columns.
df = df.drop(['Index','Month','Day','Time of Day'],axis=1)
```

```python
#Dropping null and duplicate datas.
df = df.dropna().drop_duplicates()
#Getting information from data frame.
df.info()
```

```python
#Seperating labels.
label = df.Target
#Seperating and dropping columns with object data type.
color = df.Color
source = df.Source
drop_cols = ['Target','Color','Source']
df_num = df.drop(drop_cols,axis=1)
df_num.reset_index(drop=True,inplace=True)
```

```python
for i in drop_cols :
    print(i)
    print(df[i].unique())
```

```python
#Encoding columns with object data type.
encoded_color = DataFrame(encoder.fit_transform(color),columns=['color'])
encoded_source = DataFrame(encoder.fit_transform(source),columns=['source'])
encoded_color.reset_index(drop=True,inplace=True)
encoded_source.reset_index(drop=True,inplace=True)
#Reunioning columns to a unique data frame.
ready_df = concat([df_num,encoded_color,encoded_source],axis=1)
```

```python
#Getting information from data frame .
ready_df.info()
```

```python
#Scaling datas.
scaled_df = DataFrame(scaler.fit_transform(ready_df),columns=ready_df.columns)
```

```python
#Seperating datas to train and test sets.
xtr , xts , ytr , yts = train_test_split(scaled_df,label,test_size=0.2,random_state=42)
```

```python
#Instancing from 'DecisionTreeClassifier' object and fit it with train data set.
model = DecisionTreeClassifier(max_depth=21,max_features=15).fit(xtr,ytr)
#Calculating scores.
model.score(xtr,ytr)*100
```

```python
model.score(xts,yts)*100
```
